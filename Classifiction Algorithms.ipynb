```python
x = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print("Shape of features training data:",x_train.shape)
print('Shape of features testing data:',x_test.shape)
```

```python
print("Shape of Label training data:",y_train.shape)
print('Shape of Label testing data:',y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
x_train_scale = scale.fit_transform(x_train)
x_test_scale = scale.fit_transform(x_test)
```

```python
ModelDT = DecisionTreeClassifier()

ModelDT.fit(x_train,y_train)

train_mod = ModelDT.predict(x_train)
```

```python
# Check training accuracy

from sklearn.metrics import accuracy_score
```

```python
print("Training Accuracy of Decesiontree model is:",accuracy_score(y_train,train_mod))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
list = [3,5,7,9,11,13,15]

for i in list: 
  ModelKNN = KNeighborsClassifier(n_neighbors=i)

  ModelKNN.fit(x_train_scale,y_train)

  train_mod = ModelKNN.predict(x_train_scale)

  print(f"Training Accuracy of KNeighborsClassifier model of n_neighbors is {i}:",accuracy_score(y_train,train_mod))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
list = [100,200,300,400,500]

for i in list: 
  ModelLR = LogisticRegression(max_iter=i)

  ModelLR.fit(x_train_scale,y_train)

  train_mod = ModelLR.predict(x_train_scale)

  print(f"Training Accuracy of LogisticRegression model of max_iter is {i}:",accuracy_score(y_train,train_mod))
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Define an SVM model
svm_model = SVC()

# Define hyperparameters to tune
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'],
    'class_weight': [None, 'balanced']
}

# Perform Grid Search
grid_search = GridSearchCV(svm_model, param_grid, cv=5)
grid_search.fit(x_train_scale, y_train)


# Best parameters
print("Best Hyperparameters:", grid_search.best_params_)

#train_pred = grid_search.predict(x_train_scale)

print("Best Training Score for svm:", grid_search.best_score_)

```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
list = [50,100,200,300,400,500]

for i in list: 
  ModelRF = RandomForestClassifier(n_estimators=i)

  ModelRF.fit(x_train,y_train)

  train_pred = ModelRF.predict(x_train)

  print(f"Training Accuracy of RandomForestClassifier model of n_estimators is {i}:",accuracy_score(y_train,train_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
list = [100,200,300,400,500]

for i in list: 
  ModelBG = BaggingClassifier(n_estimators=i,random_state=42)

  ModelBG.fit(x_train,y_train)

  train_pred = ModelBG.predict(x_train)

  print(f"Training Accuracy of BaggingClassifier model of n_estimators is {i}:",accuracy_score(y_train,train_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
list = [100,200,300,400,500]

for i in list: 
  ModelBS = AdaBoostClassifier(n_estimators=i,random_state=42)

  ModelBS.fit(x_train,y_train)

  train_pred = ModelBS.predict(x_train)

  print(f"Training Accuracy of AdaBoostClassifier model of n_estimators is {i}:",accuracy_score(y_train,train_pred))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
ModelVT= VotingClassifier(estimators=[
                        ("DT", DecisionTreeClassifier()),
                        ("KNN", KNeighborsClassifier(n_neighbors=5)),
                        ("LR", LogisticRegression(max_iter=200)),
                        ("SVM", SVC(probability=True)),
                        ("RFC", RandomForestClassifier(n_estimators=100)),
                        ("Bagg_clf",BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)),
                        ("Boost_clf",AdaBoostClassifier(estimator=DecisionTreeClassifier(), n_estimators=500, algorithm="SAMME", random_state=42))
                          ], voting="soft")

ModelVT.fit(x_train,y_train)

train_pred = ModelVT.predict(x_train)

print("Training Accuracy of VotingClassifier model is :",accuracy_score(y_train,train_pred))
```
